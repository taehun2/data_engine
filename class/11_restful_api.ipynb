```python
main_api = "https://www.mapquestapi.com/directions/v2/route?"
orig = "Rome, Italy"
dest = "Frascati, Italy"
key = "5XAJwG1F2glkV1jOug3FpkA9kBdEPrTG"
```

```python
url = main_api + urllib.parse.urlencode({"key":key, "from":orig, "to":dest})
url
```

```python
json_data = requests.get(url).json()
json_data
```
